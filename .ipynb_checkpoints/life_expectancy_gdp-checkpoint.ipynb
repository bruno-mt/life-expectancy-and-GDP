Let's see how GDP behaves now.

```python
graph = sns.catplot(data=df, x="Year", hue="Country", y="GDP", kind="point", aspect=1.5)
graph.set_axis_labels("Year", "GDP (USD)")
xdump = plt.xticks(rotation=90)
```

GDP seems to be all over the place for our six countries. Some rise, Germany rises first but then drops a bit and the rest seem to maintain themselves throughout the years.

This is not enough information to tell us if there is any relationship between them, so let's explore deeper.

```python
df_meanlife = df.groupby("Country")["Life expectancy at birth (years)"].mean().reset_index()
df_meanlife
```

```python
lifegraph = sns.barplot(data=df_meanlife, x="Country", y="Life expectancy at birth (years)")
xdump = plt.xticks(rotation=90)
```

This graph with the average Life Expectancy for each of the six countries show how most of them are very close, between 74 and 79 years. Zimbabwe continues being the only outlier, with an average Life Expectancy of 50 years.

```python
df_meangdp = df.groupby("Country")["GDP"].mean().reset_index()
df_meangdp
```

```python
sns.barplot(data=df_meangdp, x="Country", y="GDP")
xdump = plt.xticks(rotation=90)
```

We see that there's a huge difference in size. Chile and Zimbabwe are pretty much not visible. 

I'll try using a logarithmic scale to solve this.

```python
gdpgraph = sns.barplot(data=df_meangdp, x="Country", y="GDP", log=True)
xdump = plt.xticks(rotation=90)
```

We already knew Zimbabwe was very low on the gaph, but at least we can see it now. 

```python
df_means = df_meanlife
df_means['GDP'] = df_meangdp['GDP']
df_means
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
linegdp, = ax1.plot(df_means["Country"], df_means["GDP"], label="GDP", linewidth=2)
linelife, = ax2.plot(df_means["Country"], df_means["Life expectancy at birth (years)"], color="r", label="Life Expectancy", linewidth=2)
plt.legend([linegdp, linelife], ["GDP","Life Expectancy"])
xdump = plt.xticks(rotation=90)
plt.show()
```

This graph is comparing the behavior of the mean values for Life Expectancy and GDP from our six countries. We can see there's not a lot of similarities and it seems all over the place.

So, making these graphs didn't show any correlation. The previous graphs where we had every single year for all of our countries neither showed any relationship. 

I'll try calculating a correlation factor now and see what it tells.


```python
print(df["Life expectancy at birth (years)"].corr(df["GDP"]))
```

By calculating the Pearson correlation between Life Expectancy and GDP, we found a correlation of 0.343.

A value close to 1 or -1 indicates a near perfect linear correlation between the variables, while a value of 0 indicates there's no correlation at all. 

Our value being 0.343 indicates there's a low correlation between Life Expectancy and GDP (for the data of all countries together).

Now, let's try to calculate the correlation for each individual country.

```python
#Make dataframes for each country
df_chile = df[df["Country"] == "Chile"]
df_china = df[df["Country"] == "China"]
df_germany = df[df["Country"] == "Germany"]
df_mexico = df[df["Country"] == "Mexico"]
df_usa = df[df["Country"] == "United States of America"]
df_zimbabwe = df[df["Country"] == "Zimbabwe"]

print("Chile's correlation:")
print(df_chile["Life expectancy at birth (years)"].corr(df_chile["GDP"], method="pearson"))

print("\nChina's correlation:")
print(df_china["Life expectancy at birth (years)"].corr(df_china["GDP"]))

print("\nGermany's correlation:")
print(df_germany["Life expectancy at birth (years)"].corr(df_germany["GDP"]))

print("\nMexico's correlation:")
print(df_mexico["Life expectancy at birth (years)"].corr(df_mexico["GDP"]))

print("\nUnited States' correlation:")
print(df_usa["Life expectancy at birth (years)"].corr(df_usa["GDP"]))

print("\nZimbabwe's correlation:")
print(df_zimbabwe["Life expectancy at birth (years)"].corr(df_zimbabwe["GDP"]))
```

Calculating the individual values for each country show a high correlation between their Life Expectancy and GDP.

Let's try to scatter plot this.

```python
sns.scatterplot(data = df, x="Life expectancy at birth (years)", y="GDP", hue="Country")
```

Here we can see why, when we tried calculating a correlation between all the values, it was low. But doing so for each individual country turned out to have a high linear correlation, with China being the lowest amongst them but still having a 0.90 correlation value.


Thanks to this graphs, we are now able to answer some of the questions we originally had.

Let's remember our goals:

- Identify the relationship between the GDP and life expectancy of six countries.
- Would an increase in GDP also increase the life expectancy?
- Does the proportion between these metrics is similar for every country?
- To find if it's possible and viable/positive to have growth in both GDP and Life Expectancy and how.



**Findings:**
1. There's a high, positive linear relationship between GDP and Life Expectancy at Birth for all of our countries (from 0.90 to 0.98)
2. The increase in GDP is related to the increase in Life Expectancy. However, with this information we can't ensure this relationship is causal. Or which metric moves which. (i.e. if the increase in GDP causes an increase in Life Expectancy or the increase in Life Expectancy is the one responsible for the increase in GDP)
3. The proportion between the metrics is similar for all of these countries. From 0.90 to 0.98. Some have a higher correlation, but they're all very close together.
4. Things we can ensure are:
    - Life expectancy has increased throughout time for all of the countries. With Zimbabwe having an exceptional increase right after a small drop.
    - GDP has also increased throughout the years, with China having the biggest jump.
    - The average life expectancy for these countries is between 70 - 80 years, except for Zimbabwe, whose average Life Expectancy is around 50 years.
